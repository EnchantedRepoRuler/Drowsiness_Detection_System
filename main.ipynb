```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

Load Model

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
model
```

Make Detection with Images

```python
img = 'https://assets.bbhub.io/professional/sites/24/2017/09/india-traffic-car2.jpg'
```

```python
results = model(img)
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

Real Time Detection

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
  ret, frame = cap.read()

  # Make Detections

  results = model(frame)

  cv2.imshow('YOLO', np.squeeze(results.render()))

  if cv2.waitKey(10) & 0xFF == ord('q'):
    break
cap.release()
cv2.destroyAllWindows()
```
